```python
# Common imports
import numpy as np
import pylab as pl
import pandas as pd
```

```python
# Make plots prettyful
pl.style.use('fivethirtyeight')
```

```python
# Pokemon
pokemon = pd.read_csv('data/pokemon.csv')

# Abilities
pokemon_abilities = pd.read_csv('data/pokemon_abilities.csv')
abilities = pd.read_csv('data/abilities.csv')
pokemon_abilities = pokemon_abilities.merge(abilities[['id', 'identifier']], left_on='ability_id', right_on='id')
pokemon = pokemon.merge(
    pokemon_abilities[pokemon_abilities['slot'] == 1][['pokemon_id', 'identifier']].rename(columns={'identifier': 'ability1'}),
    how='left', left_on='id', right_on='pokemon_id'
)
pokemon = pokemon.merge(
    pokemon_abilities[pokemon_abilities['slot'] == 2][['pokemon_id', 'identifier']].rename(columns={'identifier': 'ability2'}),
    how='left', left_on='id', right_on='pokemon_id'
)
pokemon = pokemon.merge(
    pokemon_abilities[pokemon_abilities['slot'] == 3][['pokemon_id', 'identifier']].rename(columns={'identifier': 'hidden_ability'}),
    how='left', left_on='id', right_on='pokemon_id'
)

# Species
pokemon_species = pd.read_csv('data/pokemon_species.csv') 
pokemon_species = pokemon_species.merge(
    pd.read_csv('data/pokemon_colors.csv').rename(columns={'identifier': 'color', 'id': 'cid'}), 
    left_on='color_id', right_on='cid', how='left'
)
pokemon_species = pokemon_species.merge(
    pd.read_csv('data/pokemon_habitats.csv').rename(columns={'identifier': 'habitat', 'id': 'hid'}),
    left_on='habitat_id', right_on='hid', how='left'
)
pokemon_species = pokemon_species.merge(
    pd.read_csv('data/pokemon_shapes.csv').rename(columns={'identifier': 'shape', 'id': 'sid'}),
    left_on='shape_id', right_on='sid', how='left'
)
flavor = pd.read_csv('data/pokemon_species_flavor_text.csv')
pokemon_species = pokemon_species.merge(
    flavor[(flavor['language_id'] == 9) & (flavor['version_id'] == 26)].rename(columns={'species_id': 'id'}), on='id'
)
names = pd.read_csv('data/pokemon_species_names.csv')
pokemon_species = pokemon_species.merge(
    names[names['local_language_id'] == 9].rename(columns={'pokemon_species_id': 'id'}), on='id'
)
pokemon_species = pokemon_species.merge(
    names[names['local_language_id'] == 1].rename(columns={'pokemon_species_id': 'id', 'name': 'kana'})[['id', 'kana']], on='id'
)

pokemon = pokemon.merge(
    pokemon_species[['id', 'color', 'habitat', 'shape', 'flavor_text', 'genus', 'kana']].rename(
        columns={'id': 'species_id'}
    ), how='left', on='species_id'
)


# Base stats
pokemon_stats = pd.read_csv('data/pokemon_stats.csv')
stats = pd.read_csv('data/stats.csv') 
for s in stats.iloc[:6].iterrows():
    pokemon = pokemon.merge(
        pokemon_stats[pokemon_stats['stat_id'] == s[1].id].rename(
            columns={'pokemon_id': 'id', 'base_stat': s[1].identifier}
        )[['id', s[1].identifier]], on='id'
    )
    
    
# Types
pokemon_types = pd.read_csv('data/pokemon_types.csv') 
pokemon_types = pokemon_types.merge(
    pd.read_csv('data/types.csv').rename(columns={'id': 'type_id'}), on='type_id'
)
pokemon = pokemon.merge(
    pokemon_types[pokemon_types['slot'] == 1].rename(columns={'pokemon_id': 'id', 'identifier': 'type1'})[['id', 'type1']]
)
pokemon = pokemon.merge(
    pokemon_types[pokemon_types['slot'] == 2].rename(columns={'pokemon_id': 'id', 'identifier': 'type2'})[['id', 'type2']], how='left'
)


pokemon[[
    'id', 'identifier', 'kana', 'height', 'weight', 'type1', 'type2',
    'ability1', 'ability2', 'hidden_ability', 
    'color', 'habitat', 'shape', 'genus', 'flavor_text',
    'hp', 'attack', 'defense', 'special-attack', 'special-defense', 'speed'
]]
```

```python
pokemon[[
    'id', 'identifier', 'kana', 'height', 'weight', 'type1', 'type2',
    'ability1', 'ability2', 'hidden_ability', 
    'color', 'habitat', 'shape', 'genus', 'flavor_text',
    'hp', 'attack', 'defense', 'special-attack', 'special-defense', 'speed'
]].to_csv('pokemon.csv', index=False)
```
